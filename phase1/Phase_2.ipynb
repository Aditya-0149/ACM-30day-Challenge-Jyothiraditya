```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("creditcard.csv")
df
```

```python
# Data Preprocessing
df.isnull().sum()
df.drop_duplicates()
df['Class'].value_counts()
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
Scaler = StandardScaler()
X_Scaled = Scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_Scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca,y,test_size=0.3,random_state=42)
```

```python
svm_linear = SVC(kernel='linear')
svm_linear.fit(X_train,y_train)
linear_p = svm_linear.predict(X_test)

svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(X_train,y_train)
rbf_p = svm_rbf.predict(X_test)

svm_poly = SVC(kernel='poly',degree=3)
svm_poly.fit(X_train,y_train)
poly_p = svm_poly.predict(X_test)
```

```python
print("svm_linear Accuracy:",accuracy_score(y_test,linear_p))
print("svm_linear Classification report:",classification_report(y_test,linear_p, zero_division=0))
```

```python
print("svm_rbf Accuracy:",accuracy_score(y_test,rbf_p))
print("svm_rbf Classification report:",classification_report(y_test,rbf_p, zero_division=0))
```

```python
print("svm_poly Accuracy:",accuracy_score(y_test,poly_p))
print("svm_poly Classification report:",classification_report(y_test,poly_p, zero_division=0))
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=linear_p, palette='coolwarm', alpha=0.6)
plt.title("SVM with linear Kernel")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Predicted Class")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=rbf_p, palette='coolwarm', alpha=0.6)
plt.title("SVM with rbf Kernel")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Predicted Class")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=poly_p, palette='coolwarm', alpha=0.6)
plt.title("SVM with polynomial Kernel")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Predicted Class")
plt.show()
```
